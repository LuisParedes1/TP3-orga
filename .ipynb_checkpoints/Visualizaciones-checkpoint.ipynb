```python
# importacion general de librerias y de visualizacion (matplotlib y seaborn)
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

%matplotlib inline

plt.style.use('default') # haciendo los graficos un poco mas bonitos en matplotlib
#plt.rcParams['figure.figsize'] = (20, 10)

sns.set(style="whitegrid") # seteando tipo de grid en seaborn

pd.options.display.float_format = '{:20,.2f}'.format # suprimimos la notacion cientifica en los outputs

import warnings
warnings.filterwarnings('ignore')

from datetime import datetime
```

> In this competition you are predicting the probability that an online transaction is fraudulent, as denoted by the binary target isFraud.

El target es predecir si la transaccion es fraudulenta o no

```python
train_identity = pd.read_csv('./ieee-fraud-detection/train_identity.csv')

train_transaction = pd.read_csv('./ieee-fraud-detection/train_transaction.csv',
                                usecols=['TransactionID','isFraud'])
```

```python
display(train_identity)
```

```python
train_identity.columns
```

# Analiso el dataset identity

Uno ambos datasets identity con su target para ver que relaciones encuentro

```python
train_identity_with_target = train_identity.merge(train_transaction, how='outer', on='TransactionID')
```

```python
train_identity_with_target.corr()
```

## Busco correlaciones entre los labels

```python
train_identity_with_target.corr().iloc[1:-1].isFraud.plot.bar(title='Correlaciones entre labels y isFraud',color=(0.6, 0.6, 0.5, 0.3),  edgecolor='blue')
```

## Analizo id_17

Observamos que el label 'id_17' parece tener la mejor correlacion, por lo tanto graficamos un BoxPlot para ver su distribucion

```python
train_identity_with_target.boxplot('id_17','isFraud',meanline=True, rot=45, backend='matplotlib')
```

## Analizo id_22

Otro feature relevante es id_32, por ende vemos su distribucion con un violin plot


```python
cm = 1/2.54
fig, ax = plt.subplots(1,1, figsize=(35*cm, 35*cm))


sns.violinplot('isFraud','id_22', data=train_identity_with_target, ax=ax)
sns.despine()

ax.set_title("Es Fraude")
ax.set_ylabel("Cantidad id_22")
```

Es inmediato que exiten muchos mas casos de isFraud = 0 para este id, lo que nos da una idea de que la data puede estar poco distribuida

## Vista general de la data de identity

Queremos saber si la data esta distribuida

```python
heat_map_db = train_identity_with_target.groupby('isFraud').agg({'count'})

ax = sns.heatmap(heat_map_db, cmap ='Reds')
ax.set_title('Correlaciones entre features y fraude')
ax.set_xlabel ('Fraude', fontsize = 10)
ax.set_ylabel ('id', fontsize = 10)
```

Vemos que tenemos un problema de clasificacion desbalanceado. Existe una gran cantidad de transacciones que caen en la categoria no fraudulenta (0), por lo tanto no se puede entrenar el modelo con una cantidad aleatoria de registros sino que hay que ser selectivos.

## Analisis Adicional

Veo que tanta informacion pierdo al hacer drop nan

```python
cm = 1/2.54
fig, ax = plt.subplots(1,2, figsize=(25*cm, 15*cm))

ax[0].hist(train_identity_with_target.isFraud, color=(0.2, 0.3, 0.5, 0.3),  edgecolor='blue')
ax[0].set_xlabel ('isFraud', fontsize = 10)
ax[0].set_ylabel ('Cantidad', fontsize = 10)
ax[0].set_title('Cantidad de Fraude previo a limpiar data')


ax[1].hist(train_identity_with_target.dropna().isFraud, color=(0.2, 0.3, 0.5, 0.3),  edgecolor='red')
ax[1].set_xlabel ('isFraud', fontsize = 10)
ax[1].set_ylabel ('Cantidad', fontsize = 10)
ax[1].set_title('Cantidad de Fraude despues de limpiar data')
```

Vemos que no podemos hacer dropnans porque perdemos una gran cantidad de datos

## Que dispositivos se produce la mayor cantidad de fraudes?

Veamos desde que dispositivos se producen la mayor cantidad de fraude

```python
train_identity_with_target[train_identity_with_target.isFraud == 1].DeviceType.hist()
```

Y desde cuales dispositivos se produce la mayor cantidad de no fraude

```python
train_identity_with_target[train_identity_with_target.isFraud == 0].DeviceType.hist()
```

No se puede determinar mucha informacion importante, parece que los fraudes ocurren independientemente de si se accede desde el computador o desde el movil

# Analisis de la data junta

Analizo un heatmap con todos los datos unidos para ver si estaria balanceado en este caso

```python
train_identity = pd.read_csv('./ieee-fraud-detection/train_identity.csv')

train_transaction = pd.read_csv('./ieee-fraud-detection/train_transaction.csv')
```

```python
all_data = train_transaction.merge(train_identity, on='TransactionID', how='outer')
```

```python
heat_map_db = all_data.groupby('isFraud').agg({'count'})

ax = sns.heatmap(heat_map_db, cmap ='Reds')
ax.set_title('Correlaciones entre features y fraude')
ax.set_xlabel ('Feaute', fontsize = 10)
ax.set_ylabel ('EsFraude', fontsize = 10)
```

## Conclusion

Es evidente que hay un gran desbalance con toda la data junta, por ende hay que tener cuidado de no cometer overfitting y crear un modelo que solo prediga que no hay fraude nunca.
